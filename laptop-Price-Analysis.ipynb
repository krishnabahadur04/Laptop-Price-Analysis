<b>Price Distribution Plot<b>

```python
plt.figure(figsize=(15,7))
sns.histplot(df['Price_euros'],color='red',kde=True)
plt.show()
```

<b>Countplot for Categorial Value</b>

```python
def drawplot(col, df):
    plt.figure(figsize=(15, 7))
    sns.countplot(x=col, data=df, palette='plasma', hue=col)
    plt.xticks(rotation='vertical')
    plt.title(f"Countplot of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()

# Define the columns to plot
toview = ['Company', 'TypeName', 'Ram', 'OS']

# Create countplots for each column
for col in toview:
    drawplot(col, df)
```
